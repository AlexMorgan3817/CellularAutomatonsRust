```python
import numpy as np
import plotly.express as px, plotly.graph_objects as go
import pandas as pd
```

```python
# ^Test 0: (.*) \([^\)]*\): .*$ => $1
vecbench = np.array([
0.037,
0.089,
0.180,
0.332,
1.949,
0.184,
0.807,
0.008,
0.043,
1.040])
#### HashMap<(u32, u32), i32> result
hashmapbench = np.array([0.069,
0.257,
0.657,
1.245,
12.712,
0.682,
2.559,
0.031,
0.155,
5.846])
```

```python
comparison:pd.DataFrame = pd.DataFrame({"Hash": hashmapbench, "Vec": vecbench})
comparison.sort_values("Hash", inplace=True)
comparison['x/y'] = comparison.Hash / comparison.Vec
comparison
```

```python
go.Figure().add_scatter(x=[i for i in range(0, len(comparison))], y=comparison['x/y'])
```

```python
go.Figure(layout=dict(width=800)).add_traces([\
	go.Scatter(name="Hashmap", x=[i for i in range(0, len(hashmapbench))], y=hashmapbench),\
	go.Scatter(name="Vec", x=[i for i in range(0, len(vecbench))], y=vecbench)\
])
```
